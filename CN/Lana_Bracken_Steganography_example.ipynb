```python
original_image_name = "buffalo.png" # filename to hide a message in
image_name_with_message = "Lana_Bracken_buffalo_modified.png" # desired filename of the image with hidden message
message_to_hide = "On, on Buffaloes,,, we'll bring home the victory!  W-T-A-M-, WTAM, Fight! Fight! Fight!" # message you want to hide
```

```python
from stegano import lsb
secret=lsb.hide(original_image_name, message_to_hide)
```

```python
secret.save(image_name_with_message)
```

```python

hidden_message = lsb.reveal(image_name_with_message)
try:
  hidden_message = lsb.reveal("Capture.PNG")
except Exception as e:
  if "Impossible to detect message" in str(e):
    print("impossible to detect message")
# Basic check for incomplete or corrupt data:
if not hidden_message:
  print("no hidden message")
print(hidden_message)
```
